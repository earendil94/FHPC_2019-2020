```python
#Plotting theoretical speedup for serial sum parallel algorithm

P = linspace(1,2000,2000)
n = {3,4,5,6,7,8,9,10,11}
T_comp = 2*10e-9
T_read = 0
T_comm = 1.5*10e-6

def Speedup(x,r,T_comp,T_read,T_comm):
    T_P = T_read + T_comm * 2 * (x - 1) + T_comp * (x + 1 + r/x)
    T_1 = r * T_comp
    return T_1/T_P

max_proc_val = []
fh = figure(figsize=(20,20))
fh.suptitle("Theoretical Speedup of Partial Sum")
for i in n:
    ax = fh.add_subplot(3,3,i-2)
    ax.set_title("n = " + str(10**i))
    ax.set_xlabel("P")
    ax.set_ylabel("Speedup(P)")
    y = Speedup(P,10**i,T_comp,T_read,T_comm)
    max_proc_val.append(P[y.argmax()])
    plot(P,Speedup(P,10**i,T_comp,T_read,T_comm))
    
print(max_proc_val)
fh.text(.5, .05,"Theoretical speedup calculated with the above mentioned model for the computation of the partial sum from 1 to N, plotted for different values of N",ha='center')
fh.savefig("img/Theoretical_speedup_partial_sum.png")
```

```python
#Strong scaling plots
import pandas

def practical_speedup(T_1,T_P):
    return T_1/T_P

#Walltime_data
df = pandas.read_csv('csv/mpi-pi_STRONG_.csv',
                     delimiter=";",
                     float_precision='round_trip',
                    names=['Rank', 'P', 'T(P)','N'])

#ElapsedTime_data
df_e = pandas.read_csv('csv/mpi-pi_STRONG_ELAPSED.csv',
                     delimiter=";",
                     float_precision='round_trip',
                    names=['Rank', 'P', 'T(P)','N'])

#We extract the maximum walltime for every P for a fixed N
idx = df.groupby(['N','P'])['T(P)'].transform(max) == df['T(P)']
strong_df = df[idx]

#We extract the maximum elapsed for every P for a fixed N
idx = df_e.groupby(['N','P'])['T(P)'].transform(max) == df_e['T(P)']
strong_df_e = df_e[idx]


df_T1 = pandas.read_csv('csv/pi_STRONG_.csv',
                        delimiter=";",
                        names=['Rank', 'P', 'T(P)','N'])


counter = 0
fh = figure(figsize=(15,10))
fh.suptitle("Strong scaling Montecarlo Pi")
for i in df_T1.N.unique():
    ax = fh.add_subplot(2,2,counter+1)
    ax.set_xlabel("P")
    ax.set_ylabel("Speedup(P)")
    ax.set_title("n = " + str(i))
    k_df_e = strong_df_e.loc[strong_df_e['N'] == i][[ 'P','T(P)','N' ]].drop_duplicates()
    k_df = strong_df.loc[strong_df['N'] == i][[ 'P','T(P)','N' ]].drop_duplicates()
    k_df_T1 = df_T1.loc[df_T1['N'] == i]
    ps = practical_speedup(k_df_T1['T(P)'].values,k_df['T(P)'].values)
    ps_e = practical_speedup(k_df_T1['T(P)'].values,k_df_e['T(P)'].values)
    b = k_df['P'].values
    plot(b, ps, label="Walltime")
    plot(b, ps_e, label="Elapsed")
    ax.legend()
    counter += 1

fh.savefig("img/StrongScalingMontecarloPi.png")



```

```python
#Strong scaling serial sum plots
import pandas

df = pandas.read_csv('csv/mpiPartialSumSerial_STRONG.csv',
                     delimiter=";",
                     float_precision='round_trip',
                    names=['Rank', 'P', 'T(P)','N'])

df_t1 = pandas.read_csv('csv/partialSumSerial_STRONG.csv',
                       delimiter=";",
                     float_precision='round_trip',
                    names=['Bla', 'P', 'T','N'])

#print(df)
#We extract the maximum walltime for every P for a fixed N
idx = df.groupby(['N','P'])['T(P)'].transform(max) == df['T(P)']
strong_df = df[idx]


counter = 0
fh = figure(figsize=(15,15))
fh.suptitle("Strong Scaling")
for i in strong_df.N.unique():
    ax = fh.add_subplot(2,2,counter+1)
    ax.set_xlabel("P")
    ax.set_ylabel("Speedup(P)")
    ax.set_title("n = " + str(i))
    k_df = strong_df.loc[strong_df['N'] == i]
    k_df_t1 = df_t1.loc[df_t1['N'] == i ]
    speedup = practical_speedup(k_df_t1['T'].values,k_df['T(P)'].values)
    plot(k_df['P'],speedup, label="Actual speedup")
    plot(k_df['P'],k_df['P'],label="Perfect speedup")
    ax.legend()
    counter += 1
    

savefig("img/SumStrongScaling.png")
```
